```python
import pickle

import matplotlib.pyplot as plt
```

```python
pickle_file = 'data/GENKI_32_2019-12-01-22.06.46.pickle'
with open(pickle_file, "rb") as input_file:
    data = pickle.load(input_file)
```

```python
randidx = data['randidx']
pixels = data['pixels']
images = data['images']
labels = data['labels']
masks = data['masks']
```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
print(labels[4])
axs[0].imshow(images[4].reshape(256, 256), cmap='bone')
axs[1].imshow(masks[4].reshape(256, 256), cmap='bone')
axs[2].imshow(pixels[4].reshape(32, 32), cmap='seismic')

plt.show()
```

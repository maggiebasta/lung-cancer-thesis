```python
import os
import shutil

import numpy as np
import matplotlib.pyplot as plt
import skimage.transform as trans
from skimage.io import imread

import data_downloader

import warnings
warnings.filterwarnings("ignore")
```

# Download and prepare LIDC_IDRI raw dataset

```python
# Clear out old data
shutil.rmtree('data/prepared')
```

```python
data_downloader.download_and_extract(20, 50, delete_raw=True)
```

# Split into test and train

```python
# Clear out old data
shutil.rmtree('data/train')
shutil.rmtree('data/test')
```

```python
data_downloader.test_train_split('data/prepared', 'data/train', 'data/test')
```

# Inspect prepared data

### Training data

```python
train_idx = np.random.randint(len(os.listdir('data/train/image1')))
print(train_idx)

train_img_0 = imread(f'data/train/image0/{train_idx}.png')
train_img_1 = imread(f'data/train/image1/{train_idx}.png')
train_img_2 = imread(f'data/train/image2/{train_idx}.png')
train_img_3 = imread(f'data/train/image3/{train_idx}.png')

train_mask_0 =  imread(f'data/train/label0/{train_idx}.png')
train_mask_1 =  imread(f'data/train/label1/{train_idx}.png')
train_mask_2 =  imread(f'data/train/label2/{train_idx}.png')
train_mask_3 =  imread(f'data/train/label3/{train_idx}.png')



fig, axs = plt.subplots(2, 4, figsize=(20, 8))
axs[0][0].imshow(train_img_0, cmap='bone')
axs[0][1].imshow(train_img_1, cmap='bone')
axs[0][2].imshow(train_img_2, cmap='bone')
axs[0][3].imshow(train_img_3, cmap='bone')

axs[1][0].imshow(train_mask_0, cmap='bone')
axs[1][1].imshow(train_mask_1, cmap='bone')
axs[1][2].imshow(train_mask_2, cmap='bone')
axs[1][3].imshow(train_mask_3, cmap='bone')


plt.show()
```


### Test data

```python
test_idx = np.random.randint(len(os.listdir('data/test/image1')))
print(test_idx)

test_img_0 = imread(f'data/test/image0/{test_idx}.png')
test_img_1 = imread(f'data/test/image1/{test_idx}.png')
test_img_2 = imread(f'data/test/image2/{test_idx}.png')
test_img_3 = imread(f'data/test/image3/{test_idx}.png')


test_mask_0 =  imread(f'data/test/label0/{test_idx}.png')
test_mask_1 =  imread(f'data/test/label1/{test_idx}.png')
test_mask_2 =  imread(f'data/test/label2/{test_idx}.png')
test_mask_3 =  imread(f'data/test/label3/{test_idx}.png')


fig, axs = plt.subplots(2, 4, figsize=(20, 8))
axs[0][0].imshow(test_img_0, cmap='bone')
axs[0][1].imshow(test_img_1, cmap='bone')
axs[0][2].imshow(test_img_2, cmap='bone')
axs[0][3].imshow(test_img_3, cmap='bone')

axs[1][0].imshow(test_mask_0, cmap='bone')
axs[1][1].imshow(test_mask_1, cmap='bone')
axs[1][2].imshow(test_mask_2, cmap='bone')
axs[1][3].imshow(test_mask_3, cmap='bone')


plt.show()
```

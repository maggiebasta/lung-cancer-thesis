```python
import math 
import os

import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pydicom
import scipy.ndimage as ndimage

from skimage.filters import gabor
from skimage.morphology import watershed
```

```python
# key = 'LIDC-IDRI/LIDC-IDRI-0001/01-01-2000-30178/3000566-03192/000003.dcm'
key = 'LIDC-IDRI/LIDC-IDRI-0002/01-01-2000-98329/3000522-04919/000001.dcm'
ds = pydicom.dcmread(key)
img = ds.pixel_array
plt.imshow(img, cmap=plt.cm.bone) 
plt.show()
```

# Smoothing

```python
img_smooth = ndimage.gaussian_filter(img, sigma=1)
plt.imshow(img_smooth, cmap=plt.cm.bone) 
plt.show()
```

```python
img_filt_real, img_filt_imag = gabor(img_smooth, frequency=0.6)
plt.imshow(img_filt_real, cmap=plt.cm.bone) 
plt.show()
```

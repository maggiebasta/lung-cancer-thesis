# Inspect prepared data

### Training data

```python
train_idx = np.random.randint(len(os.listdir('data/train/image')))

train_img = Image.open(f'data/train/image/{train_idx}.png') 
train_mask = Image.open(f'data/train/label/{train_idx}.png') 

print(train_idx)

fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].imshow(np.array(train_img), cmap='gray')
axs[1].imshow(np.array(train_mask), cmap='gray')
plt.show()
```

### Test data

```python
test_idx = np.random.randint(len(os.listdir('data/test/image')))

test_img = Image.open(f'data/test/image/{test_idx}.png') 
test_mask = Image.open(f'data/test/label/{test_idx}.png') 

print(test_idx)

fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].imshow(np.array(test_img), cmap='bone')
axs[1].imshow(np.array(test_mask), cmap='bone')
plt.show()
```
